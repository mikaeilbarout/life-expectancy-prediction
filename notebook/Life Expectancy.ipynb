```python
def change_ (x):
    if x=='Unknown':
        x=np.NaN
    return x
```

```python
file=file.applymap(change_)
```

```python
file
```

```python
file1=pd.get_dummies(file['Status'],drop_first=True)
file1
```

```python
file2=pd.get_dummies(file['Continent'],drop_first=True)
file2
```

```python
file_org=pd.concat([file,file1,file2],axis=1)
file_org
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(file_org.isnull(),cbar=False)
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(file_org.corr(),annot=True,cbar=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(file_org.drop(['Country','Continent','Status','Life_expectancy ','Year'],axis=1),
                                               file_org['Life_expectancy '],random_state=101,test_size=0.3)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
pd.DataFrame(lr.coef_,X_train.columns,columns=['coef'])
```

```python
fig , ax=plt.subplots()
ax.scatter(x=y_test,y=pred,color='r')
ax.set_xlabel('y test')
ax.set_ylabel('Predicted y')
plt.tight_layout()
```

```python
sns.distplot(pred-y_test,bins=30,color='r')
```
